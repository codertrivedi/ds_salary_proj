```python
import pandas as pd
```

```python
df = pd.read_csv('glassdoor_jobs.csv')
```

```python
df
```

```python
#salary parsing
```

```python
df = df[df['Salary Estimate'] != '-1']
```

```python
df
```

```python
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
```

```python
salary
```

```python
minus_Kd = salary.apply(lambda x: x.replace('K', '').replace('$',''))
```

```python
minus_Kd
```

```python
df['hourly'] = df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0) 
```

```python
df
```

```python
df['employer-provided'] = df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary:' in x.lower() else 0) 
```

```python
df
```

```python
min_hr = minus_Kd.apply(lambda x: x.lower().replace('per hour', '').replace('employer provided salary',''))
```

```python
df['min salary'] = min_hr.apply(lambda x: int(x.split('-')[0]))
df['max salary'] = min_hr.apply(lambda x: int(x.split('-')[1]))
df['avg_salary'] = (df.min_salary+df.max_salary)/2
```

```python
#company name text only
```

```python
df['company_txt'] = df.apply(lambda x: x['Company Name'] if x['Rating'] <0 else x['Company Name'][:-3], axis=1)
```

```python
df
```

```python
#state field
```

```python
df['job_state'] = df['Location'].apply(lambda x: x.split(',')[1])
```

```python
df.job_state.value_counts()
```

```python
df['job_state'] = df.apply(lambda x: 1 if x.Location==x.Headquarters else 0, axis = 1)
```
